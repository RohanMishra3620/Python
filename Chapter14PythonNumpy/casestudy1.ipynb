Case Study-Insurance claim Records using Numpy

Consider an insurance Claim Records using Numpy

```python
import numpy as np
claims=np.array( [[12000,15000,18000],
[10000,13000,17000],
[9000,11000,16000]]
)
print("Insurace claim Data (INR): ")
claims
```

```python
# Access claims amount of policyholder 3 for the first month
print(f"Claim for policyholder 3 in month 1: {claims[2,0]}")
# Access claims amount of policyholder 1 for the second month
print(f"Claim for policyholder 1 in month 2: {claims[0,1]}")
```

```python
# Total claims per policyholder (row-wise sum)
print("Total claims per policyholder:", np.sum(claims, axis=1))

# Average claims per claim type (column-wise mean)
print("Average claims per claim type:", np.mean(claims, axis=0).round(2))
```

```python
# Indentify High-value claims(above INR 15,000)
# Find claims greater than 150000
print(f"High-value Claims {claims[claims>15000]}")
```
