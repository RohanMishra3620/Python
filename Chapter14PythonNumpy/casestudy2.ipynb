Case Study : Customer Feedback Analysis in Retail Industry Using Numpy

Overview :
A retail chain collection Feedback through surveys .The dataset contains:
*Numerical rating (1-5): indicating satifaction level
*Textual Feedback : Comments from customer

Objective :
*Import  and process mixed data(number & text)using Numpy
*Analyize customer rating statistically
*Extract text comments for qualitive analysis

```python
import numpy as np

data = np.array([
    ["C001", 5, "Excellent service and friendly staff"],
    ["C002", 4, "Good product selection but long wait time"],
    ["C003", 3, "Average experience checkout was slow"],
    ["C004", 5, "Loved the discounts! Will shop again"],
    ["C005", 2, "Poor customer support not satisfied"],
    ["C006", 4, "Good quality but a bit expensive"],
    ["C007", 1, "Terrible experience rude staff"],
    ["C008", 5, "Amazing deals and fast service"]
])

print(data)
```

```python
ratings=data[:,1].astype(int)
print(f"Total Feedback Entries : {len(ratings)}")
print(f"Highest rating {np.max(ratings)}")
print(f"Lowest rating {np.min(ratings)}")
print(f"Average rating {np.mean(ratings):.2f}")
```

```python
unique_ratings,rating_count=np.unique(ratings,return_counts=True)
print("\nCustomer feedback Analyis summary")
print()
print("**Rating Distribution**")
for rating,count in zip(unique_ratings,rating_count):
    print(f"{rating} :{count} customer")
```

```python
# Extract text feedback for Qualitative Insight
comments=data[:,2]
print("\n**sample customer feedback comments**")
print()
for com in comments:
    print(f" -{com}")
```
