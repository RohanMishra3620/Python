 (a) Give 4 ways of creating a series.

```python
import pandas as pd
df=pd.Series([1,2,3,4])
print(df)
df=pd.Series({101:"Rohan",102:"Ravi"})
print(df)
df=pd.Series(range(1,6))
print(df)
df=pd.Series()
import numpy as np
df=pd.Series(np.array([1,2,3,4,5]))
df
```
