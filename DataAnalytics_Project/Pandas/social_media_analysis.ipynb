🔹 SECTION E — Engagement Behavior & User Psychology

21.	How many posts have likes above the overall average?

22.	How many posts have comments greater than likes?

23.	Identify posts with:

•	High engagement but low views

24.	Identify posts with:

•	High views but low engagement

25.	What proportion of posts fall into each category above?


```python
print(f"{len(df[df['likes'] > df['likes'].mean()])} posts have likes above the average")
print(f"{len(df[df['comments'] > df['likes']])} posts have more comments than likes")
high_eng_low_views = df[
    (df['engagement_rate'] > df['engagement_rate'].mean()) &
    (df['views'] < df['views'].mean())
]

low_eng_high_views = df[
    (df['engagement_rate'] < df['engagement_rate'].mean()) &
    (df['views'] > df['views'].mean())
]

print(f"High-engagement & low-views posts: {len(high_eng_low_views)}")
print(f"Low-engagement & high-views posts: {len(low_eng_high_views)}")

print("\nHigh engagement but low views posts:")
print(high_eng_low_views)

print("\nLow engagement but high views posts:")
print(low_eng_high_views)

total_posts = len(df)

# 1. Likes > average likes
likes_above_avg_prop = (
    len(df[df['likes'] > df['likes'].mean()]) / total_posts
)

# 2. Comments > likes
comments_gt_likes_prop = (
    len(df[df['comments'] > df['likes']]) / total_posts
)

# 3. High engagement & low views
high_eng_low_views_prop = (
    len(df[
        (df['engagement_rate'] > df['engagement_rate'].mean()) &
        (df['views'] < df['views'].mean())
    ]) / total_posts
)

# 4. Low engagement & high views
low_eng_high_views_prop = (
    len(df[
        (df['engagement_rate'] < df['engagement_rate'].mean()) &
        (df['views'] > df['views'].mean())
    ]) / total_posts
)

print("===== PROPORTION OF POSTS IN EACH CATEGORY =====")
print(f"Likes > average likes: {likes_above_avg_prop:.2%}")
print(f"Comments > likes: {comments_gt_likes_prop:.2%}")
print(f"High engagement & low views: {high_eng_low_views_prop:.2%}")
print(f"Low engagement & high views: {low_eng_high_views_prop:.2%}")

```

SECTION F — Length-Based Content Strategy (Advanced)

26.	Categorize posts into:Short (<50),Medium (50–100),Long (>100)
27.	Compare average engagement_rate across these categories
28.	Which length category performs best?
29.	Does longer content guarantee better engagement?


```python
short=df[df["post_length"]<50]
print(short)
medium = df[(df["post_length"] >= 50) & (df["post_length"] <= 100)]
print(medium)
long=df[df["post_length"]>100]
print(long)


vs = pd.Series({
    "Short": short["engagement_rate"].mean().round(4),
    "Medium": medium["engagement_rate"].mean().round(4),
    "Long": long["engagement_rate"].mean().round(4)
}, name="avg_engagement_rate").to_frame()




print(vs)

best_category = vs["avg_engagement_rate"].idxmax()
best_value = vs["avg_engagement_rate"].max()

print(f"Best performing length category: {best_category} ({best_value:.4f})")

# Final insight
print("Conclusion:")
print("Longer content does not guarantee better engagement.")
print(f"{best_category} length posts perform best based on average engagement rate.")

```

🔹 SECTION H — Relationship & Trend Analysis

35.	Relationship between views and engagement_rate

36.	Correlation between:

•	likes & engagement_rate

•	shares & engagement_rate

37.	Does higher reach guarantee higher engagement?


```python
# Value	Meaning
# +1	Perfect positive correlation
# 0	No correlation
# -1	Perfect negative correlation

# Correlation between views & engagement rate
corr_views = df["views"].corr(df["engagement_rate"])
print(f"Correlation between views and engagement rate: {corr_views:.3f}")

# Correlation between likes & engagement rate
corr_likes = df["likes"].corr(df["engagement_rate"])
print(f"Correlation between likes and engagement rate: {corr_likes:.3f}")

# Correlation between shares & engagement rate
corr_shares = df["shares"].corr(df["engagement_rate"])
print(f"Correlation between shares and engagement rate: {corr_shares:.3f}")

# No, higher reach does not guarantee higher engagement.
```

Engagement rate varies significantly across platforms, indicating uneven user interaction.

One platform consistently delivers the highest engagement, making it ideal for engagement-focused campaigns.

Video content performs better than text and image posts in terms of engagement and shareability.

Short and medium-length posts generally achieve higher engagement than long posts.
