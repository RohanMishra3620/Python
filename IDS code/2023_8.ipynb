Insert 2 new records in DataFrame.

Add a new column for percentage of marks of both semesters.

```python
df["percentage"]=(df["Semester1"]+df["Semester2"])/11
df
```

Add a new column for grades based on percentage:


Percentage	Grade
≥ 90	O
≥ 75 to < 90	A+
≥ 60 to < 75	A
≥ 50 to < 60	B+
≥ 40 to < 50	B
< 40	F

```python
def gradefun(p):
    if p >= 90:
        return "O"
    elif p >= 75:
        return "A+"
    elif p >= 60:
        return "A"
    elif p >= 50:
        return "B+"
    elif p >= 40:
        return "B"
    else:
        return "F"
df["Grade"]=df["percentage"].apply(gradefun)
df
```
