```python
import pandas as pd
data={
    "INFY":[5.2,4.8,5.5,5.0,4.9,5.3],
    "RELIANCE":[8.5,12.2,7.8,15.3,10.1,9.0],
    "HDFCBANK":[6.1,5.5,6.7,6.2,5.8,6.0]
}
df=pd.DataFrame(data)
df
```

```python
print("Mean Monthly return\n ",df.mean())
```

```python
print("\n Standard deviation(Risk)")
print(df.std())
```

```python
import matplotlib.pyplot as plt
df.plot(kind="box")
plt.ylabel("Monthly")
plt.grid()
```
