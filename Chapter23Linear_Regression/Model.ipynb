Design and implement a simple linear regression model to predict Sunday newspaper circulation based on weekly newspaper circulation.

1. smf.ols()

OLS means:

Ordinary Least Squares

Ye linear regression model banata hai.

2. "sunday ~ daily"

Formula format:

y∼x

Yahan:

sunday → dependent variable (target)
daily → independent variable (input)

Meaning:

Predict sunday circulation using daily circulation.

3. data=df_news

Batata hai ki data kis dataframe me stored hai.

Yahan:

df_news

dataset ka naam hai.

4. .fit()

Model ko train karta hai.

Matlab:

regression line calculate karega
slope/intercept find karega
best fit equation banayega

```python
import statsmodels.formula.api as smf

model = smf.ols("sunday ~ daily", data=df_news).fit()
```

```python
model.params
```

sunday=13.83+1.33(weekly)

```python
model.pvalues
```

predict using model

```python
newdata=pd.Series([200,300,500])
```

```python
data_pred=pd.DataFrame(newdata,columns=["daily"])
```

```python
data_pred
```

```python
data_pred["Sunday"]=model.predict(data_pred)
data_pred
```

```python
model.rsquared
```
