Linear Regression Overview

Linear Regression is a supervised learning algorithm used to model the relationship between one or more independent variables (features) and a continuous dependent variable (target).

Types of Linear Regression:
1. Simple (or Single) Linear Regression:

Uses one independent variable to predict a continuous target variable by fitting a straight line.

2. Multiple Linear Regression:

Uses two or more independent variables to predict the target by fitting a multi-dimensional hyperplane.
