(a) Write Python code to create the following series:

Name	Marks

Harsh	95

Rahul	85

Amit	75

Anita	45

Anjali	80

Amita	90

Yashasvi	65


```python
import pandas as pd

df = pd.Series(
    data=[95, 85, 75, 45, 80, 90, 65],
    index=["Harsh", "Rahul", "Amit", "Anita", "Anjali", "Amita", "Yashasvi"]
)
df.name = "Marks"          # Column name
df.index.name = "Student" 

print(df)
```

(b) Show details of ‘Amit’.



```python
print(df["Amit"])
```
