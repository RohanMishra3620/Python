# Social Media Data Analytics using NumPy

This project analyzes social media post performance 
across different platforms and content types.


```python
import numpy as np
```

```python
data = np.genfromtxt(
    "../Data/Social_dataset.csv",
    delimiter=",",
    names=True,          
    dtype=None,
    encoding="utf-8"
)
```

```python
# 🔹 A. Descriptive Analytics (Daily Analyst Work)
#1. Total number of posts
total_number_post=len(data)
print(f"Total number post = {total_number_post}")
#2. Average views, likes, comments, shares
avg_views = np.mean(data["views"])
print(f"Average views are {avg_views:.2f}")
avg_likes = np.mean(data["likes"])
print(f"Average likes are {avg_likes:.2f}")
avg_comments = np.mean(data["comments"])
print(f"Average comments are {avg_comments:.2f}")
avg_shares = np.mean(data["shares"])
print(f"Average shares are {avg_shares:.2f}")
#3. Maximum & minimum views
maximum_views=np.max(data["views"])
print(f"Maximum views are : {maximum_views}")
minumum_views=np.min(data["views"])
print(f"Minumum views are : {minumum_views}")
#4. Median engagement rate shows typical engagement excluding outliers

median_engagement_rate = np.median(data["engagement_rate"])
print(f"Median engagemet are : {median_engagement_rate:.5f}")

# 5.	Standard deviation of likes
standard_likes=np.std(data["likes"])
print(f"Standard deviation of likes : {standard_likes:.2f}")
```

```python
# 🔹 B. Platform Analysis 
# 6.	Platform-wise:
# Average views
# Average engagement_rate

facebook_mask = data["platform"] == "Facebook"
twitter_mask = data["platform"] == "Twitter"
instagram_mask = data["platform"] == "Instagram"
print(f"Average views on Facebook: {np.mean(data["views"][facebook_mask]):.2f}")
print(f"Average views on Twitter: {np.mean(data["views"][twitter_mask]):.2f}")
print(f"Average views on Instagram: {np.mean(data["views"][instagram_mask]):.2f}")
print(f"Average engagement_rate on Facebook: {np.mean(data["engagement_rate"][facebook_mask]):.2f}")
print(f"Average engagement_rate on Twitter: {np.mean(data["engagement_rate"][twitter_mask]):.2f}")
print(f"Average engagement_rate on Instagram: {np.mean(data["engagement_rate"][instagram_mask]):.2f}")
# Instagram has higher average engagement rate compared to Facebook and Twitter
# Facebook generates higher average views but lower engagement
```

```python
# 7.	Platform with:
# 	Highest average likes
# 	Lowest average engagement

platforms = np.array(["Facebook", "Twitter", "Instagram"])


avg_likes_platform_wise = np.array([
    np.mean(data["likes"][facebook_mask]),
    np.mean(data["likes"][twitter_mask]),
    np.mean(data["likes"][instagram_mask])
])
#argmax gives the index of highest value
highest_likes_index = np.argmax(avg_likes_platform_wise)
print(f"Platform with highest average likes: {platforms[highest_likes_index]}")


avg_engagement_platform_wise = np.array([
    np.mean(data["engagement_rate"][facebook_mask]),
    np.mean(data["engagement_rate"][twitter_mask]),
    np.mean(data["engagement_rate"][instagram_mask])
])

lowest_engagement_index = np.argmin(avg_engagement_platform_wise)
print(f"Platform with lowest average engagement: {platforms[lowest_engagement_index]}")


# 8.	Facebook vs Instagram:
# Kaunsa zyada engagement deta hai?

if(np.mean(data["engagement_rate"][facebook_mask])>np.mean(data["engagement_rate"][instagram_mask])):
    print("Facebook is more engagement rate")
else:
    print("Instagram is more engagement rate") 
```

```python
# 9. Post_type-wise average engagement_rate

print(set(data["post_type"]))

video_mask = data["post_type"] == "Video"
text_mask  = data["post_type"] == "Text"
image_mask = data["post_type"] == "Image"

print(f"Average engagement rate for Video posts: {np.mean(data['engagement_rate'][video_mask]):.2f}")
print(f"Average engagement rate for Text posts: {np.mean(data['engagement_rate'][text_mask]):.2f}")
print(f"Average engagement rate for Image posts: {np.mean(data['engagement_rate'][image_mask]):.2f}")

# 10.	Best performing post_type
post_type=np.array(["Video","Text","Image"])
avg_engagement_post_wise=np.array([
np.mean(data['engagement_rate'][video_mask]),
np.mean(data['engagement_rate'][text_mask]),
np.mean(data['engagement_rate'][image_mask])
]
)

print(f"Best performing post_type is : {post_type[np.argmax(avg_engagement_post_wise)]}")

# 11.	Worst performing post_type

print(f"Worst performing post_type is : {post_type[np.argmin(avg_engagement_post_wise)]}")

# 12.	Video posts ka:
# Avg views
# Avg shares
print(f"Average views for Video posts: {np.mean(data['views'][video_mask]):.2f}")
print(f"Average shares for Video posts:{np.mean(data['shares'][video_mask]):.2f}")
```

🔹 D. Engagement Behavior (REAL Analyst Thinking 🔥)

```python
# 13.	Top 5 posts by engagement_rate
np.sort(data["engagement_rate"])[-5:]
```

```python
# 14. Posts with likes greater than average likes
likes_above_avg_mask = data["likes"] > avg_likes

print("Posts with likes greater than average likes:")
print(data[likes_above_avg_mask])


# 15. Posts where comments are greater than likes
comments_more_than_likes_mask = data["comments"] > data["likes"]

print("\nPosts where comments are greater than likes:")
print(data[comments_more_than_likes_mask])
```

```python
# 16.	High-engagement low-views posts identify karo
mask=(data["likes"]<avg_likes) & (data["engagement_rate"]>np.mean(data["engagement_rate"]))

print("High-engagement but low-views posts:")
print(data[mask])
```

```python
# E. Strategy-Based Questions (Company Uses This 💼)
# 17.	Short posts (<50 length) vs Long posts (>100)
import pandas as pd
short_post=data["post_length"]<50
long_post=data["post_length"]>100

Short_vs_longPost  = {
    "Properties": ["views", "likes", "comments","shares","engagement_rate"],
    "Short_post": [np.mean(data["views"][short_post]),
                   np.mean(data["likes"][short_post]), 
                   np.mean(data["comments"][short_post]),
                   np.mean(data["shares"][short_post]),
                   np.mean(data["engagement_rate"][short_post])],
    "Long_post": [np.mean(data["views"][long_post]), 
                  np.mean(data["likes"][long_post]),
                  np.mean(data["comments"][long_post]),
                  np.mean(data["shares"][long_post]),
                  np.mean(data["engagement_rate"][long_post])
                  ]
}
answer=pd.DataFrame(Short_vs_longPost)
answer.round(2)
```

```python
# Platform-wise total shares (same order as platforms array)
sum_share_platform_wise = np.array([
    np.sum(data["shares"][facebook_mask]),
    np.sum(data["shares"][twitter_mask]),
    np.sum(data["shares"][instagram_mask])
])

most_shareable_index = np.argmax(sum_share_platform_wise)

print(f"Platform with most shareable content: {platforms[most_shareable_index]}")
```

```python
# 19.	Engagement_rate aur views ka relation check karo

ER_vs_Views = {
   "Properties": ["mean","max","min"],
   "Engagement_rate": [np.mean(data["engagement_rate"]),
                  np.max(data["engagement_rate"]), 
                  np.min(data["engagement_rate"])
                  ],
   "Views": [np.mean(data["views"]), 
                 np.max(data["views"]),
                 np.min(data["views"])
                 ]
}
ans=pd.DataFrame(ER_vs_Views)
 
ans
print(ans.round(2))
print()
correlation = np.corrcoef(data["views"], data["engagement_rate"])[0, 1]

print(f"Correlation between views and engagement rate: {correlation:.2f}")
# | Correlation Value | Meaning                  |
# | ----------------- | ------------------------ |
# | +1 ke paas        | Strong positive relation |
# | 0 ke paas         | No relation              |
# | -1 ke paas        | Negative relation        |

```

```python
# 20. Platform engagement ke liye weak hai?

mean_ER_platform_wise = np.array([
    np.mean(data["engagement_rate"][facebook_mask]),
    np.mean(data["engagement_rate"][twitter_mask]),
    np.mean(data["engagement_rate"][instagram_mask])
])

weak_platform_index = np.argmin(mean_ER_platform_wise)

print(f"Platform with weakest engagement: {platforms[weak_platform_index]}")
print()
print("\n===== END OF ANALYSIS =====")
```
